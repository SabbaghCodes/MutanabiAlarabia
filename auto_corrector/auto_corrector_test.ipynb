```python
import json
import sys
sys.path.append(r'/content/drive/MyDrive/Colab Notebooks/mutanabi_alarabia/auto_corrector') 

from auto_corrector import AutoCorrection
```

```python
def wiki_data():
    data_list = []
    file_data = open("/content/drive/MyDrive/Colab Notebooks/mutanabi_alarabia/auto_corrector/ar_arz_wiki_corpus.json")
    for data in file_data:
        data_list.append((json.loads(data))['text'])
    txt = " ".join(data_list)
    return txt.split()
```

```python
words_lst = wiki_data()
ar_letters = "ابتثجحخدذرزسشصضطظعغفقكلمنويءؤئإألإ"
auto_correct = AutoCorrection(words_lst, ar_letters)
```

```python
while True:
    txt = input('Enter your sentence: ')
    txt = txt.split()
    sentence = ""
    for word in txt:
        suggestion_word = auto_correct.get_correction(word)
        if len(suggestion_word) != 0:
            print(f"Suggestion for {word} is {suggestion_word[0][0]}")
            sentence += (suggestion_word[0][0] + " ")
        else:
            print(f"Suggestion for {word} is {word}")
            sentence += (word + " ")
    print("Full sentence after applying suggestions: ", sentence)
    log_out = input('Log out(y/n):')
    if log_out == 'y' or log_out == 'Y':
        break
```

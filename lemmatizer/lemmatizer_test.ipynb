```python
import pandas as pd
```

```python
data = pd.ExcelFile('/content/drive/MyDrive/Colab Notebooks/mutanabi_alarabia/lemmatizer/data.xlsx')
```

```python
data = pd.read_excel(data)
data
```

```python
!pip install pyarabic
import pyarabic.araby as araby

def strip_diacritics(string):
  return araby.strip_diacritics(str(string))

data['stripped_lemma'] = data.lemma.apply(strip_diacritics)

data
```

```python
word_data = data[['root', 'lemma', 'stripped_lemma', 'part_of_speech']]
word_data
```

```python
lemma_to_root_dict = pd.Series(word_data.root.values,index=word_data.stripped_lemma.values).to_dict()
```

```python
def strip_diacritics(string):
  return araby.strip_diacritics(str(string))


def lemma_to_root(lemma):
  lemma = strip_diacritics(lemma)
  lemmas = word_data.stripped_lemma[word_data.stripped_lemma.str.find(lemma)>=0]
  if len(lemmas) != 0:
    lemma_index = lemmas.index[0]
    lemma_root = word_data.iloc[lemma_index, 0]
    return lemma_root
  return None

def lemma_to_part_of_speech(lemma):
  lemma = strip_diacritics(lemma)
  lemmas = word_data.stripped_lemma[word_data.stripped_lemma.str.find(lemma)>=0]
  if len(lemmas) != 0:
    lemma_index = lemmas.index[0]
    lemma_part_of_speech = word_data.iloc[lemma_index, 3]
    if lemma_part_of_speech == 'n':
      return 'اسم (n)'
    elif lemma_part_of_speech == 'v':
      return 'فعل (v)'
    elif lemma_part_of_speech == 'a':
      return 'صفة (a)'
  return None
```

```python
print("الكلمة: إنهاء")
print("الجذر: ",lemma_to_root("إنهاء"))
print("قسم الكلام: ", lemma_to_part_of_speech("إنهاء"))

print("-----------------")

print("الكلمة: إختتم")
print("الجذر: ", lemma_to_root("إختتم"))
print("قسم الكلام: ", lemma_to_part_of_speech("إختتم"))

print("-----------------")

print("الكلمة: إستماع")
print("الجذر: ",lemma_to_root("إستماع"))
print("قسم الكلام: ", lemma_to_part_of_speech("إستماع"))

print("-----------------")

print("الكلمة: إستنتاجي")
print("الجذر: ",lemma_to_root("إستنتاجي"))
print("قسم الكلام: ", lemma_to_part_of_speech("إستنتاجي"))


print("-----------------")

print("الكلمة: إختفاء")
print("الجذر: ",lemma_to_root("إختفاء"))
print("قسم الكلام: ", lemma_to_part_of_speech("إختفاء"))
```

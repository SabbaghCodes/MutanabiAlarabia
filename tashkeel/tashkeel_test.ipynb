```python
import sys
sys.path.append(r'/content/drive/MyDrive/Colab Notebooks/mutanabi/tashkeel') 

from tashkeel import Tashkeel
```

```python
input_text = "فإن لم يكونا كذلك أتى بما يقتضيه الحال وهذا أولى"
# or
# input_text = input("الرجاء ادخال النص المراد تشكيله:  ")
```

```python
folder_location = r'/content/drive/MyDrive/Colab Notebooks/mutanabi/tashkeel'
tashkeel = Tashkeel(folder_location)
```

```python
print("preparing input")
input_int = tashkeel.prepare_input(input_text)
print("finished preparing input")

print("starting with the model")
model, graph = tashkeel.get_model()
logits = model.predict(input_int)[0]
predicted_harakat = tashkeel.logits_to_text(logits)
print("finished successfully")
```

```python
final_output = tashkeel.get_final_text(input_text, predicted_harakat)
print(final_output)
```
